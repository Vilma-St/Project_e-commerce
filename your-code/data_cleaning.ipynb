```python
# Investigating product categories in English. There are 71 categories, which makes it
# difficult to visualize and draw conclusions. 

sorted(list(product_category.product_category_name_english.unique()))
```

```python
# Reducing the amount of categories. It is important to keep in mind that the grouping 
# is subjective. 

dictionary = {'art':'arts_and_craftmanship', 'books_general_interest':'books',
              'books_imported':'books', 'books_technical':'books','cds_dvds_musicals':'audio', 
              'construction_tools_construction': 'construction', 
              'construction_tools_lights':'construction',
              'construction_tools_safety':'construction', 
              'costruction_tools_garden':'construction',
              'costruction_tools_tools':'construction','diapers_and_hygiene':'baby',
              'drinks':'food_drink', 'food':'food_drink', 
              'fashio_female_clothing':'fashion', 'fashion_bags_accessories':'fashion',
              'fashion_childrens_clothes':'fashion', 'fashion_male_clothing':'fashion',
              'fashion_shoes':'fashion','fashion_sport':'fashion',
              'fashion_underwear_beach':'fashion', 'furniture_bedroom':'furniture',
              'furniture_decor':'furniture', 'furniture_living_room':'furniture',
              'furniture_mattress_and_upholstery':'furniture', 
              'kitchen_dining_laundry_garden_furniture':'furniture',
              'home_appliances':'home_comfort', 'home_appliances_2':'home_comfort',
              'home_comfort_2':'home_comfort', 'home_confort':'home_comfort',
              'home_construction':'home_comfort', 
              'signaling_and_security':'security_and_services',
              'small_appliances_home_oven_and_coffee':'small_appliances'}


lst = []
for i in product_category['product_category_name_english']:
    if i in (dictionary.keys()):
        lst.append(dictionary[i])
    else:
        lst.append(i)

product_category['product_category_name_english'] = lst

len(list(product_category.product_category_name_english.unique()))

```
